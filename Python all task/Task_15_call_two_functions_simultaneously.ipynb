```python
import threading
import time

def a():
    while True:
        print("Function a is running")
        time.sleep(1)

def b():
    while True:
        print("Function b is running")
        time.sleep(1)

# Create threads for functions a and b
thread_a = threading.Thread(target=a)
thread_b = threading.Thread(target=b)

# Start the threads
thread_a.start()
thread_b.start()

thread_a.join()
thread_b.join()
```
